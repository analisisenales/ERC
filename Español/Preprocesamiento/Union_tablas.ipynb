# Unión

Ahora que ya tenemos los datasets en forma de csv y ademas lo suficientemente depurados, se procede a unirlos para generar una tabla mas grande

```python
import pandas as pd

# 1. Cargar los dos archivos CSV
df_biopro = pd.read_csv('P_BIOPRO_nombres_claros.csv')
df_demo = pd.read_csv('P_DEMO_nombres_claros.csv')

# 2. Unir los archivos usando 'ID_Participante' como llave
# El parámetro how='inner' conserva solo a los participantes que están en ambos archivos
df_final = pd.merge(df_biopro, df_demo, on='ID_Participante', how='inner')

# 3. Limpieza: Eliminar filas que tengan valores nulos tras la unión
df_final_limpio = df_final.dropna()

# 4. Guardar el archivo unificado
nombre_salida = 'Unificacion_tablas.csv'
df_final_limpio.to_csv(nombre_salida, index=False, encoding='utf-8-sig')

# Reporte rápido en consola
print("=== Reporte de la Unión ===")
print(f"✅ Archivo guardado con éxito como: '{nombre_salida}'")
print(f"🔹 Filas totales (participantes completos): {len(df_final_limpio)}")
print(f"🔹 Columnas totales combinadas: {df_final_limpio.shape[1]}")
```
