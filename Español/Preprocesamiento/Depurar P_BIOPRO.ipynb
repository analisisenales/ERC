# Depuracion 

En la tabla "P_BIOPRO.xpt" se repiten columnas con los aributos, pero sus unidades de medida son las que cambian, siendo este cambio en factores de 100 o 1000, lo que quiere decir que este puede estar en mmg/dL o mmol/L, lo cual solo afecta el valor numerico pero no el analisis en sí mismo. Las causas por las que se modificara la tabla son:

<ol>
    <li>Se eliminan las columnas que tengan como unico dato al valor nulo</li>
    <li>Se eliminan las columnas que posean mismos atributos pero con distinta unidad de medida, conservando aquellas que posean el SI</li>
    <li>Se eliminan las filas que como unico dato tengan el SEQN</li>
</ol>

Con esta idea se logra una mejor depuracion de los datos

```python
import pandas as pd

# 1. Cargar el archivo .xpt
nombre_entrada = 'P_BIOPRO.xpt'
df = pd.read_sas(nombre_entrada)

# --- PASO A: ELIMINAR COLUMNAS ESPECÍFICAS ---
columnas_a_eliminar = [
    'LBXSAL', 'LBXSBU', 'LBXSCR', 'LBXSGB', 'LBXSGL', 'LBXSIR', 'LBXSPH', 'LBXSTB', 'LBXSCA', 'LBXSCH',
    'LBXSTP', 'LBXSTR', 'LBXSUA', 'LBDSATLC', 'LBDSTBLC', 'LBDSGTLC'
]
df_limpio = df.drop(columns=columnas_a_eliminar, errors='ignore')

# --- PASO A.1: ELIMINAR COLUMNAS SI TODA LA COLUMNA ES NULA ---
# axis=1 indica columnas, how='all' indica que TODOS los valores deben ser NaN
df_limpio = df_limpio.dropna(axis=1, how='all')

# --- PASO B: ELIMINAR FILAS VACÍAS ---
# Definimos qué columnas usaremos para evaluar si la fila está "vacía"
# (Usamos las columnas actuales de df_limpio)
columnas_datos = [col for col in df_limpio.columns if col != 'SEQN']

# how="all" significa que solo borra la fila si TODAS las columnas de datos son NaN
df_limpio = df_limpio.dropna(subset=columnas_datos, how='all')

# --- PASO C: GUARDAR RESULTADOS ---
nombre_salida_csv = 'P_BIOPRO_limpio.csv'
df_limpio.to_csv(nombre_salida_csv, index=False, encoding='utf-8-sig')

# --- REPORTE DE CAMBIOS ---
print("-" * 30)
print("REPORTE DE LIMPIEZA FINAL")
print("-" * 30)
print(f"Filas originales: {len(df)}")
print(f"Filas resultantes: {len(df_limpio)}")
print(f"Columnas originales: {len(df.columns)}")
print(f"Columnas eliminadas por lista: {len(columnas_a_eliminar)}")
print(f"Columnas finales: {len(df_limpio.columns)}")
print(f"Archivo guardado como: {nombre_salida_csv}")
```

Inicialmente eran 10409 filas de datos y 41 columnas, pero luego de la depuracion se tienen 9477 filas con 25 columnas, con datos efectivos para poder trabajarlos sin ningun problema. Al final arroja una nueva tabla llamada "P_BIOPRO_limpio.csv". Algo importante a observar es que la extension del archivo cambio a uno mas amigable

# Cambio a nombres mas legibles 

Ahora, se cambian los nombres de los atributos a unos mas legibles

```python
# 1. Diccionario de mapeo específico para las columnas que conservaste (unidades SI)
mapping_nombres = {
    'SEQN': 'ID_Participante',
    'LBXSATSI': 'ALT_U_L (Alanina Aminotransferasa)',
    'LBDSALSI': 'Albumina_g_L',
    'LBXSAPSI': 'Fosfatasa_Alcalina_U_L',
    'LBXSASSI': 'AST_U_L (Aspartato Aminotransferasa)',
    'LBXSC3SI': 'Bicarbonato_mmol_L',
    'LBDSBUSI': 'Nitrogeno_Ureico_mmol_L (BUN)',
    'LBXSCLSI': 'Cloruro_mmol_L',
    'LBXSCK': 'Creatina_Quinasa_U_L',
    'LBDSCRSI': 'Creatinina_umol_L',
    'LBDSGBSI': 'Globulina_g_L',
    'LBDSGLSI': 'Glucosa_mmol_L',
    'LBXSGTSI': 'GGT_U_L (Gamma-glutamil Transaminasa)',
    'LBDSIRSI': 'Hierro_umol_L',
    'LBXSLDSI': 'LDH_U_L (Lactato Deshidrogenasa)',
    'LBXSOSSI': 'Osmolalidad_mmol_Kg',
    'LBDSPHSI': 'Fosforo_mmol_L',
    'LBXSKSI': 'Potasio_mmol_L',
    'LBXSNASI': 'Sodio_mmol_L',
    'LBDSTBSI': 'Bilirrubina_Total_umol_L',
    'LBDSCASI': 'Calcio_Total_mmol_L',
    'LBDSCHSI': 'Colesterol_mmol_L',
    'LBDSTPSI': 'Proteinas_Totales_g_L',
    'LBDSTRSI': 'Trigliceridos_mmol_L',
    'LBDSUASI': 'Acido_Urico_umol_L'
}

# 2. Renombrar las columnas en el DataFrame limpio
df_p_biopro_final = df_limpio.rename(columns=mapping_nombres)

# 3. Exportar a un nuevo archivo CSV
nombre_archivo_humano = 'P_BIOPRO_nombres_claros.csv'
df_p_biopro_final.to_csv(nombre_archivo_humano, index=False, encoding='utf-8-sig')

print(f"✅ Proceso completado. El archivo '{nombre_archivo_humano}' ha sido generado con nombres legibles.")
print(f"Total de columnas procesadas: {len(df_p_biopro_final.columns)}")
```
