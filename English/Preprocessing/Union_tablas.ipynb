# Merge

Now that the datasets have been converted into CSV format and sufficiently cleaned, the next step is to merge them in order to generate a larger and more comprehensive dataset.

```python
import pandas as pd

# 1. Load the two CSV files
df_biopro = pd.read_csv('P_BIOPRO_nombres_claros.csv')
df_demo = pd.read_csv('P_DEMO_nombres_claros.csv')

# 2. Merge the files using 'Participant_ID' as the key
# The how='inner' parameter retains only participants present in both files
df_final = pd.merge(df_biopro, df_demo, on='Participant_ID', how='inner')

# 3. Cleaning: Remove rows containing null values after the merge
df_final_cleaned = df_final.dropna()

# 4. Save the unified file
output_name = 'Unificacion_tablas.csv'
df_final_cleaned.to_csv(output_name, index=False, encoding='utf-8-sig')

# Quick report in console
print("=== Merge Report ===")
print(f"✅ File successfully saved as: '{output_name}'")
print(f"🔹 Total rows (complete participants): {len(df_final_cleaned)}")
print(f"🔹 Total combined columns: {df_final_cleaned.shape[1]}")
```
