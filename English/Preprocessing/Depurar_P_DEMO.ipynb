# Data Cleaning

In the "P_DEMO.xpt" dataset, some columns contain repeated text-based comments, making them unsuitable for analysis since numerical data is required. Therefore, the dataset will be modified by removing these columns to improve data processing and analysis.

Additionally, the removal of rows containing only null values, except for the SEQN identifier, represents a data cleaning step that prepares the file for further processing with the other classifiers.


```python
import pandas as pd

# 1. Load the .xpt file
input_name = 'P_DEMO.xpt'
df = pd.read_sas(input_name)

# --- STEP A: REMOVE SPECIFIC COLUMNS ---
columns_to_remove = [
    'SDDSRVYR', 'RIDSTATR', 'RIDAGEMN',
       'RIDRETH1', 'RIDRETH3', 'RIDEXMON', 'DMDBORN4', 'DMDYRUSZ', 'DMDEDUC2',
       'DMDMARTZ', 'RIDEXPRG', 'SIALANG', 'SIAPROXY', 'SIAINTRP', 'FIALANG',
       'FIAPROXY', 'FIAINTRP', 'MIALANG', 'MIAPROXY', 'MIAINTRP', 'AIALANGA',
       'WTINTPRP', 'WTMECPRP', 'SDMVPSU', 'SDMVSTRA', 'INDFMPIR'
]
df_limpio = df.drop(columns=columns_to_remove, errors='ignore')

# --- STEP A.1: REMOVE COLUMNS IF THE ENTIRE COLUMN IS NULL ---
# axis=1 indicates columns, how='all' indicates that ALL values must be NaN
df_limpio = df_limpio.dropna(axis=1, how='all')

# --- STEP B: REMOVE EMPTY ROWS ---
# Define which columns will be used to evaluate whether the row is "empty"
# (Using the current columns of df_limpio)
data_columns = [col for col in df_limpio.columns if col != 'SEQN']

# how="all" means that a row is removed only if ALL data columns are NaN
df_limpio = df_limpio.dropna(subset=data_columns, how='all')

# --- STEP C: SAVE RESULTS ---
output_csv_name = 'P_DEMO_limpio.csv'
df_limpio.to_csv(output_csv_name, index=False, encoding='utf-8-sig')

# --- CHANGE REPORT ---
print("-" * 30)
print("FINAL CLEANING REPORT")
print("-" * 30)
print(f"Original rows: {len(df)}")
print(f"Resulting rows: {len(df_limpio)}")
print(f"Original columns: {len(df.columns)}")
print(f"Columns removed by list: {len(columns_to_remove)}")
print(f"Final columns: {len(df_limpio.columns)}")
print(f"File saved as: {output_csv_name}")
```

Initially, the dataset contained 15,560 rows and 29 columns. After the data cleaning process, the dataset was reduced to 15,560 rows and 3 columns, containing effective data suitable for further analysis without any issues. The process generates a new dataset named "P_DEMO_limpio.csv". An important aspect to note is that the file extension was changed to a more accessible and user-friendly format.

# Renaming to More Readable Names

Ahora, se cambian los nombres de los atributos a unos mas legibles para comodidad de estudio.

```python
import numpy as np

# 1. Mapping dictionary
mapping_demo = {
    'SEQN': 'Participant_ID',
    'SDDSRVYR': 'Data release cycle',
    'RIDSTATR': 'Interview/Examination status',
    'RIAGENDR': 'Gender',
    'RIDAGEYR': 'Age',
    'RIDAGEMN': 'Age in months at screening - 0 to 24 mos',
    'RIDRETH1': 'Race/Hispanic origin',
    'RIDRETH3': 'Race/Hispanic origin w/ NH Asian',
    'RIDEXMON': 'Six-month time period',
    'DMDBORN4': 'Country of birth',
    'DMDYRUSZ': 'Length of time in US',
    'DMDEDUC2': 'Education level - Adults 20+',
    'DMDMARTZ': 'Marital status',
    'RIDEXPRG': 'Pregnancy status at exam',
    'SIALANG': 'Language of SP Interview',
    'SIAPROXY': 'Proxy used in SP Interview?',
    'SIAINTRP': 'Interpreter used in SP Interview?',
    'FIALANG': 'Language of Family Interview',
    'FIAPROXY': 'Proxy used in Family Interview?',
    'FIAINTRP': 'Interpreter used in Family Interview?',
    'MIALANG': 'Language of MEC Interview',
    'MIAPROXY': 'Proxy used in MEC Interview?',
    'MIAINTRP': 'Interpreter used in MEC Interview?',
    'AIALANGA': 'Language of ACASI Interview',
    'WTINTPRP': 'Full sample interview weight',
    'WTMECPRP': 'Full sample MEC exam weight',
    'SDMVPSU': 'Masked variance pseudo-PSU',
    'SDMVSTRA': 'Masked variance pseudo-stratum',
    'INDFMPIR': 'Ratio of family income to poverty'
}

# 2. Rename the columns
df_p_demo_final = df_p_demo_limpio.rename(columns=mapping_demo)

# --- AGE TREATMENT ---
# First: Convert individuals younger than 1 year to NaN
df_p_demo_final.loc[df_p_demo_final['Age'] < 1, 'Age'] = np.nan

# Second: Remove all rows where 'Age' is NaN
# (This removes original missing values and newly created ones)
df_p_demo_sin_nulos = df_p_demo_final.dropna(subset=['Age'])
# ---------------------------

# 3. Export to a new CSV file
demo_file_name = 'P_DEMO_nombres_claros.csv'
df_p_demo_sin_nulos.to_csv(demo_file_name, index=False, encoding='utf-8-sig')

# Quick verification
removed_rows = len(df_p_demo_final) - len(df_p_demo_sin_nulos)
print(f"✅ Process completed.")
print(f"Removed rows (Age < 1 or missing): {removed_rows}")
print(f"Total final records: {len(df_p_demo_sin_nulos)}")
```
