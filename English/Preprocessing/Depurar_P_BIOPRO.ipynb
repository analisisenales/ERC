# Data Cleaning

In the "P_BIOPRO.xpt" dataset, there are columns with repeated attributes but different measurement units. This variation occurs by factors of 100 or 1000, indicating that the values may be expressed in "mg/dL" or "mmol/L". This only affects the numerical representation of the values and does not alter the analysis itself.

The dataset will be modified according to the following criteria:

<ol>
    <li>Columns containing only null values are removed.</li>
    <li>Columns with the same attributes but different measurement units are removed, retaining those expressed in SI units.</li>
    <li>Rows containing only the SEQN value are removed.</li>
</ol>

This approach enables a more effective data cleaning process and improves the overall quality of the dataset.

```python
import pandas as pd

# 1. Load the .xpt file
input_file = 'P_BIOPRO.xpt'
df = pd.read_sas(input_file)

# --- STEP A: REMOVE SPECIFIC COLUMNS ---
columns_to_remove = [
    'LBXSAL', 'LBXSBU', 'LBXSCR', 'LBXSGB', 'LBXSGL', 'LBXSIR', 'LBXSPH', 'LBXSTB', 'LBXSCA', 'LBXSCH',
    'LBXSTP', 'LBXSTR', 'LBXSUA', 'LBDSATLC', 'LBDSTBLC', 'LBDSGTLC'
]
clean_df = df.drop(columns=columns_to_remove, errors='ignore')

# --- STEP A.1: REMOVE COLUMNS IF THE ENTIRE COLUMN IS NULL ---
# axis=1 indicates columns, how='all' indicates that ALL values must be NaN
clean_df = clean_df.dropna(axis=1, how='all')

# --- STEP B: REMOVE EMPTY ROWS ---
# Define which columns will be used to evaluate whether the row is "empty"
# (Using the current columns of clean_df)
data_columns = [col for col in clean_df.columns if col != 'SEQN']

# how="all" means that a row is removed only if ALL data columns are NaN
clean_df = clean_df.dropna(subset=data_columns, how='all')

# --- STEP C: SAVE RESULTS ---
output_file_csv = 'P_BIOPRO_clean.csv'
clean_df.to_csv(output_file_csv, index=False, encoding='utf-8-sig')

# --- CHANGE REPORT ---
print("-" * 30)
print("FINAL CLEANING REPORT")
print("-" * 30)
print(f"Original rows: {len(df)}")
print(f"Resulting rows: {len(clean_df)}")
print(f"Original columns: {len(df.columns)}")
print(f"Columns removed by list: {len(columns_to_remove)}")
print(f"Final columns: {len(clean_df.columns)}")
print(f"File saved as: {output_file_csv}")
```

Initially, the dataset contained 10,409 rows and 41 columns. After the data cleaning process, the dataset was reduced to 9,477 rows and 25 columns, containing effective data suitable for further analysis without any issues. The process generates a new table named "P_BIOPRO_clean.csv". An important aspect to note is that the file extension was changed to a more accessible and user-friendly format.

# Renaming to More Readable Names

The attribute names are now changed to more readable and descriptive names to facilitate the analysis and improve the overall usability of the dataset.

```python
# 1. Specific mapping dictionary for the retained columns (SI units)
mapping_names = {
    'SEQN': 'Participant_ID',
    'LBXSATSI': 'Alanine Aminotransferase (ALT) (U/L)',
    'LBDSALSI': 'Albumin, refrigerated serum (g/L)',
    'LBXSAPSI': 'Alkaline Phosphatase (ALP) (IU/L)',
    'LBXSASSI': 'Aspartate Aminotransferase (AST) (U/L)',
    'LBXSC3SI': 'Bicarbonate (mmol/L)',
    'LBDSBUSI': 'Blood Urea Nitrogen (mmol/L)',
    'LBXSCLSI': 'Chloride (mmol/L)',
    'LBXSCK': 'Creatine Phosphokinase (CPK) (IU/L)',
    'LBDSCRSI': 'Creatinine, refrigerated serum (umol/L)',
    'LBDSGBSI': 'Globulin (g/L)',
    'LBDSGLSI': 'Glucose, refrigerated serum (mmol/L)',
    'LBXSGTSI': 'Gamma Glutamyl Transferase (GGT) (IU/L)',
    'LBDSIRSI': 'Iron, refrigerated serum (umol/L)',
    'LBXSLDSI': 'Lactate Dehydrogenase (LDH) (IU/L)',
    'LBXSOSSI': 'Osmolality (mmol/Kg)',
    'LBDSPHSI': 'Phosphorus (mmol/L)',
    'LBXSKSI': 'Potassium (mmol/L)',
    'LBXSNASI': 'Sodium (mmol/L)',
    'LBDSTBSI': 'Total Bilirubin (umol/L)',
    'LBDSCASI': 'Total Calcium (mmol/L)',
    'LBDSCHSI': 'Cholesterol, refrigerated serum (mmol/L)',
    'LBDSTPSI': 'Total Protein (g/L)',
    'LBDSTRSI': 'Triglycerides, refrig serum (mmol/L)',
    'LBDSUASI': 'Uric acid (umol/L)'
}

# 2. Rename the columns in the cleaned DataFrame
df_p_biopro_final = clean_df.rename(columns=mapping_names)

# 3. Export to a new CSV file
new_file_name = 'P_BIOPRO_nombres_claros.csv'
df_p_biopro_final.to_csv(new_file_name, index=False, encoding='utf-8-sig')

print(f"✅ Process completed. The file '{new_file_name}' has been generated with readable names.")
print(f"Total processed columns: {len(df_p_biopro_final.columns)}")
```
